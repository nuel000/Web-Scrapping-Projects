```python


url = 'https://www.airbnb.com/s/New-York--NY/homes?query=New+York%2C+NY&checkin=&checkout=&adults=1&children=0&refinement_paths%5B%5D=%2Fhomes'

r  = requests.get(url)

soup= BeautifulSoup(r.text, 'lxml')





#Creating List to append data after scraping

Names = []  #-----> Names
Desc =[]  #-----> Description
Revs =[]  #-----> Reviews


for i in range(1,14):
    

    # Getting tags for next page button
    
    
    np = soup.find('a', class_="_1bfat5l l1j9v1wn dir dir-ltr").get('href')
    cnp = 'https://www.airbnb.com'+np # to complete url we needed to add the host name

    url = cnp

    r = requests.get(url)

    soup = BeautifulSoup(r.text, 'lxml')



    #Extracting Names

    name = soup.find_all('div', class_ = 't1jojoys dir dir-ltr')
    for i in name:
        n = i.text
        Names.append(n)



    #Extracting Description

    Des= soup.find_all('div',class_= 'nquyp1l s1cjsi4j dir dir-ltr')

    for i in Des:
        d = i.text
        Desc.append(d)




    #Extracting Reviews

    reviews = soup.find_all('span', class_ = 'r1dxllyb dir dir-ltr')


    for i in reviews:
        r = i.text
        Revs.append(r)



df =  pd.DataFrame({'Names':Names, 'Description':Desc})
    
df
```
